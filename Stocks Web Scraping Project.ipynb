```python
# Import the Webscraping libraries

from bs4 import BeautifulSoup
import requests
```

```python
# Connect to Website

url = 'https://finance.yahoo.com/most-active?offset=0&count=100'

headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:109.0) Gecko/20100101 Firefox/117.0"}

page = requests.get(url, headers=headers)

soup1 = BeautifulSoup(page.content,"html.parser")
#print(soup1)

soup2 = BeautifulSoup(soup1.prettify(),"html.parser")
#print(soup2)
```

```python
table = soup2.find('tbody')
#print(table)
```

```python
# Creating Column Headers

titles = soup2.find_all('th')
table_titles = [title.text.strip() for title in titles]

print(table_titles)
```

```python
# Import Pandas ibrary

import pandas as pd
pd.set_option('display.max_rows', 100)
```

```python
df = pd.DataFrame(columns = table_titles)
df
```

```python
rows = table.find_all('tr')
```

```python
# Supplying data into our table

for row in rows:
    row_data = row.find_all('td')
    individual_row_data = [data.text.strip() for data in row_data]
    
    length = len(df)
    df.loc[length] = individual_row_data
```

```python
# Drop Column

df = df.drop(columns = '52 Week Range')
df
```

```python
# Export data to a CSV file

df.to_csv(r"C:\Users\lpate\OneDrive\Desktop\Stocks.csv")
```
